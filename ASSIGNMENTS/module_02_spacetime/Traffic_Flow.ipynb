```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
####Input Parameters####
v_max=80.    #km/hr, Max limit of traffic velocity as traffic density approaches zero
init=10.
init_bc=10.
L=11.        #km, Length of road
rho_max=250. #cars/km, Maximum traffic density, bumper-to-bumper
nx=51       #devaitions, Resolution of grid
dt=0.001     #hours, temporal deviations
dx=L/nx      #km
####Convert to Proper Units####
v_max=v_max*1000./3600.   #meters/second
L=L*1000.                 #meters
rho_max=rho_max/1000.     #cars/meter
dt=dt*3600.               #seconds
dx=dx*1000.               #meters

km_m=1000./3600.
```

#Simulation Results: Part A

```python
####Initial Conditions for Traffic Density####
x=numpy.linspace(0,L,nx)         #Create non-dimensional array
rho0=numpy.ones(nx)*init/1000     #
rho0[10:20]=50./1000.            #cars/meter Initial conditions

####Plotting Initial Conditions####
pyplot.plot(x,rho0,color='#003366',ls='--',lw=3)
pyplot.xlim(0,L);
pyplot.ylim(0,rho_max);
```

####1) Minimum Velocity at t=0 minutes

```python
v_1=[]
for i in rho0:
    v_1.append(v_max*(1-rho0/rho_max))
v_1_min=numpy.min(v_1)
print("%.2f"%v_1_min,'m/s')
```

####2) Average Velocity at t=3 minutes

```python
#Reinitialize initial conditions (re-running this cell progresses in time beyond 3min without reinitilizing each time)
rho0=numpy.ones(nx)*init/1000     #
rho0[10:20]=50./1000.            #cars/meter Initial conditions

t_2=3.*60.       #seconds, current time 
nt_2=int(t_2/dt+1)   #deviations, temporal deviations
c=v_max*dt/dx
rho_bc=init_bc/1000.

for n in range(1,nt_2):
    rho_n=rho0.copy()
    for i in range(1,nx):
        rho_n[0]=rho_bc
        rho0[i]=rho_n[i]-c*(rho_n[i]*(1-rho_n[i]/rho_max)-rho_n[i-1]*(1-rho_n[i-1]/rho_max))

pyplot.plot(x,rho0,color='#003366',ls='--',lw=3)
pyplot.xlim(0,L);
pyplot.ylim(0,0.06);

v_2=numpy.zeros_like(rho0,dtype=float)
for i in range(nx):
    v_2[i]=v_max*(1-rho0[i]/rho_max)
v_2_av=numpy.average(v_2)
print("%.2f"%v_2_av,'m/s')
```

```python
####Minimum Velocity at t=3 minutes
####This result is used in Part B####
v_2=[]
for i in rho0:
    v_2.append(v_max*(1-rho0/rho_max))
v_2_min=numpy.min(v_2)
print("%.2f"%v_2_min,'m/s')
```

####3) Minimum Velocity at t=6 minutes

```python
#Reinitialize initial conditions
rho0=numpy.ones(nx)*init/1000     #
rho0[10:20]=50./1000.            #cars/meter Initial conditions

t_3=6.*60.       #seconds, current time 
nt_3=int(t_3/dt+1)   #deviations, temporal deviations

for n in range(1,nt_3):
    rho_n=rho0.copy()
    for i in range(1,nx):
        rho_n[0]=rho_bc
        rho0[i]=rho_n[i]-c*(rho_n[i]*(1-rho_n[i]/rho_max)-rho_n[i-1]*(1-rho_n[i-1]/rho_max))

pyplot.plot(x,rho0,color='#003366',ls='--',lw=3)
pyplot.xlim(0,L);
pyplot.ylim(0,0.06);

v_3=[]
for i in rho0:
    v_3.append(v_max*(1-rho0/rho_max))
v_3_min=numpy.min(v_3)
print("%.2f"%v_3_min,'m/s')
```
